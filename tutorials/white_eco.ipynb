```python
import b2sim as b2
from copy import deepcopy as dc
```

```python
rounds = b2.Rounds(0.1)

farms = [
    b2.initFarm(rounds.getRoundFromTime(4.99),[2,0,0])
]

buy_queue = [

    #Buy cannonship
    [b2.buyDefense(900, min_buy_time=rounds.getTimeFromRound(6))],

    #Upgrade our current farm to 320 
    [b2.upgradeFarm(0,upgrades=(3,0,0))],
    [b2.upgradeFarm(0,upgrades=(3,2,0))]
]

eco_queue_0 = [
    b2.ecoSend(time=rounds.getTimeFromRound(5), send_name='Spaced Whites')
]

eco_queue_1 = [
    b2.ecoSend(time=rounds.getTimeFromRound(5), send_name='Spaced Whites'),
    b2.ecoSend(time=rounds.getTimeFromRound(6), send_name='Spaced Blacks')
]
```

```python
initial_state_game = {
    'Cash': 200,
    'Eco': 450,
    'Eco Queue': eco_queue_0,
    'Rounds': rounds,
    'Game Round': 6,
    'Farms': farms,
    'Buy Queue': buy_queue
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 9)
game_state.viewCashEcoHistory()
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
b2.writeLog(game_state.logs, filename = 'farming_white_eco')
```
