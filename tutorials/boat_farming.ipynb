## Case 1: 2100 eco before R20, eco + boat farms

We allow the player to eco continuously throughout the farming process

```python
rounds = b2.Rounds(0.1)

buy_queue = [
    
    #Buy first boat farm
    [b2.buyBoatFarm(buffer = 0, min_buy_time = 0)],
    [b2.upgradeBoatFarm(0)],

    #Buy second boat farm
    [b2.buyBoatFarm(buffer = 0, min_buy_time = 0)],
    [b2.upgradeBoatFarm(1)],

    #Buy third boat farm
    [b2.buyBoatFarm(buffer = 0, min_buy_time = 0)],
    [b2.upgradeBoatFarm(2)],

    #Buy fourth boat farm
    [b2.buyBoatFarm(buffer = 0, min_buy_time = 0)],
    [b2.upgradeBoatFarm(3)],

    #Upgrade the first boat farm into Tempire
    [b2.upgradeBoatFarm(0)]
]

eco_queue = [
    b2.ecoSend(time=1, send_name='Grouped Pinks'),
    b2.ecoSend(time=rounds.getRoundFromTime(26), send_name='Zero')
]
```

```python
initial_state_game = {
    'Cash': 1250,
    'Eco': 2100,
    'Eco Queue': eco_queue,
    'Rounds': rounds,
    'Game Round': 19,
    'Buy Queue': buy_queue
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 34)
game_state.viewCashEcoHistory()
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
b2.writeLog(game_state.logs, filename = 'boat_farming_w_eco')
```
