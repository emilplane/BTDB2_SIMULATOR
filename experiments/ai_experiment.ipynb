```python
import b2sim.engine as b2
import b2sim.analysis as dd
```

## Experiment 1: Rounds 14 - 20

```python
rounds = b2.Rounds(0.1)

farms = [
    b2.initFarm(rounds.getTimeFromRound(7), upgrades = [3,2,0]),
    b2.initFarm(rounds.getTimeFromRound(13.9), upgrades = [3,2,0])
]

initial_state_game = {
    'Cash': 0,
    'Eco': 800,
    'Rounds': rounds, #Determines the lengths of the rounds in the game state
    'Farms': farms,
    'Game Round': 13.99
}

target_time = rounds.getTimeFromRound(20)
```

```python
winner_net = dd.run(dd.config_path, initial_state_game, target_time, num_generations=50)
winning_game_state = dd.simulate(winner_net, b2.GameState(initial_state_game), target_time)
dd.viewHistory(winning_game_state)
```

## Experiment 2: R21 - R25

```python
rounds = b2.Rounds(0.1)

farms = [
    b2.initFarm(rounds.getTimeFromRound(19.99), upgrades = [4,2,0])
]

initial_state_game = {
    'Cash': 0,
    'Eco': 1500,
    'Rounds': rounds, #Determines the lengths of the rounds in the game state
    'Farms': farms,
    'Game Round': 19.99
}

target_time = rounds.getTimeFromRound(25)
```
