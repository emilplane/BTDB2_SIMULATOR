```python
import b2sim as b2
```

```python
rounds = b2.Rounds(0.1)

farms = [
    b2.initFarm(rounds.getTimeFromRound(5), upgrades = [2,0,0]),
    b2.initFarm(rounds.getTimeFromRound(12.3), upgrades = [2,3,0])
]

buy_queue = [
    [b2.withdrawBank(1, min_buy_time=rounds.getTimeFromRound(16.99))], #Withdraw from bank before MOABs
]

initial_state_game = {
    'Cash': 0,
    'Eco': 630,
    'Eco Send': b2.ecoSend(send_name = 'Grouped Blacks'),
    'Buy Queue': buy_queue,
    'Rounds': rounds, #Determines the lengths of the rounds in the game state
    'Farms': farms,
    'Game Round': 13
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 17)
game_state.viewCashEcoHistory()
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
b2.writeLog(game_state.logs, filename = 'bank_test')
```
