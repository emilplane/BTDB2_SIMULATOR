```python
import b2sim.engine as b2
import b2sim.analysis as dd
```

```python
rounds = b2.Rounds(0.1)
farms = []

eco_queue = [
    b2.ecoSend(send_name='Spaced Whites', time = rounds.getTimeFromRound(5)),
    b2.ecoSend(send_name='Spaced Blacks', time = rounds.getTimeFromRound(6)),
    b2.ecoSend(send_name='Grouped Yellows', time = rounds.getTimeFromRound(7))
]

initial_state_game = {
    'Cash': 500,
    'Eco': 375,
    'Eco Queue': eco_queue,
    'Rounds': rounds, #Determines the lengths of the rounds in the game state
    'Game Round': 5
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 8)
dd.viewHistory(game_state)
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
b2.writeLog(game_state.logs, filename = 'eco_exp')
```

```python
rounds = b2.Rounds(0.1)

eco_queue = [
    b2.ecoSend(send_name='Grouped Reds', time = rounds.getTimeFromRound(3)),
    b2.ecoSend(send_name='Spaced Pinks', time = rounds.getTimeFromRound(4.5)),
    b2.ecoSend(send_name='Grouped Greens', time = rounds.getTimeFromRound(5))
]

buy_queue = [
    [b2.buyDefense(1100, min_buy_time=rounds.getTimeFromRound(4))]
]

initial_state_game = {
    'Cash': 0,
    'Eco': 315,
    'Eco Queue': eco_queue,
    'Buy Queue': buy_queue,
    'Rounds': rounds, #Determines the lengths of the rounds in the game state
    'Game Round': 3
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 6)
dd.viewHistory(game_state)
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
b2.writeLog(game_state.logs, filename = 'eco_exp')
```
