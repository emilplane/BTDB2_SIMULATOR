```python
import b2sim.engine as b2
import b2sim.analysis as dd
```

```python
rounds = b2.Rounds(0.1) #Set the stall factor to 0.5 for all rounds except round 20.

farms = [
    b2.initFarm(rounds.getTimeFromRound(13.9), upgrades = [2,0,0]),
    b2.initFarm(rounds.getTimeFromRound(13.9), upgrades = [2,3,0])
]

# buy_queue = [

#     #Buy a 204 farm
#     [b2.buyFarm(buffer = 250)],
#     [b2.upgradeFarm(3,upgrades=(2,0,4), buffer = 250)],

#     #As soon as we can afford to do so, sell the first three farms, and upgrade the 4th one to a Monkey Wall Street
#     [b2.upgradeFarm(3,2, min_buy_time = rounds.getTimeFromRound(24) - 1, auto_sell=3)]
# ]

buy_queue = [
    [b2.withdrawBank(1, min_buy_time=rounds.getTimeFromRound(18))]
]

eco_queue = [
    b2.ecoSend(rounds.getTimeFromRound(14), 'Grouped Blacks')
]

initial_state_game = {
    'Cash': 0,
    'Eco': 800,
    'Eco Send': b2.ecoSend(send_name='Zero'),
    'Rounds': rounds,
    'Game Round': 13.9,
    'Farms': farms,
    'Eco Queue': eco_queue,
    'Buy Queue': buy_queue
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 18.1)
dd.viewHistory(game_state)
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
b2.writeLog(game_state.logs, filename = 'mawo_flowchart', path = 'logs/')
```
