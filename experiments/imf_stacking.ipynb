```python
import b2sim as b2
```

```python
rounds = b2.Rounds(0.1)

farms = [
    b2.initFarm(rounds.getTimeFromRound(24), upgrades = [2,4,0]),
    b2.initFarm(rounds.getTimeFromRound(24), upgrades = [2,4,0]),
    b2.initFarm(rounds.getTimeFromRound(24), upgrades = [2,4,0]),
    b2.initFarm(rounds.getTimeFromRound(24), upgrades = [2,4,0]),
    b2.initFarm(rounds.getTimeFromRound(24), upgrades = [2,4,0]),
    b2.initFarm(rounds.getTimeFromRound(24), upgrades = [2,4,0])
]

buy_queue = [
    [b2.activateIMF(i) for i in range(len(farms))],
    [b2.activateIMF(i) for i in range(len(farms))],
    [b2.activateIMF(i) for i in range(len(farms))],
    [b2.activateIMF(i) for i in range(len(farms))],
    [b2.activateIMF(i) for i in range(len(farms))]
]

initial_state_game = {
    'Cash': 0,
    'Eco': 700,
    'Loan': 69000,
    'Eco Send': b2.ecoSend(send_name = 'Zero'),
    'Buy Queue': buy_queue,
    'Rounds': rounds, #Determines the lengths of the rounds in the game state
    'Farms': farms,
    'Game Round': 24.5
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 30)
game_state.viewCashEcoHistory()
print("The current cash and eco and debt are (%s,%s, %s)"%(game_state.cash, game_state.eco, game_state.loan))
b2.writeLog(game_state.logs, filename = 'imf_test')
```
