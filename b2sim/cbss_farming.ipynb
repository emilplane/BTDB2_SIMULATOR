```python
import b2sim as b2
```

```python
rounds = b2.Rounds([(0,0.25), (20,0.1)])

buy_queue = [
    [b2.buyDefense(-4250), b2.buySupplyDrop()], #Buy the supply drop, assuming we already have a 230 sniper on screen
    [b2.buyEliteSniper(0)], #Update the sniper farm at index 0 to an elite sniper
    [b2.repeatedlyBuySupplyDrops(max_buy_time = rounds.getTimeFromRound(26))] #Repeatedly buy supply drops until the specified end time
]

eco_queue = [
    b2.ecoSend(rounds.getTimeFromRound(19.2), 'Zero'),
    b2.ecoSend(rounds.getTimeFromRound(20), 'Grouped Pinks')
]

initial_state_game = {
    'Cash': 0,
    'Eco': 850,
    'Eco Send': b2.ecoSend(send_name = 'Grouped Blacks'),
    'Rounds': rounds,
    'Game Round': 13,
    'Buy Queue': buy_queue,
    'Eco Queue': eco_queue
}

game_state = b2.GameState(initial_state_game)
game_state.fastForward(target_round = 26)
game_state.viewCashEcoHistory((24,12))
print("The current cash and eco are (%s,%s)"%(game_state.cash, game_state.eco))
print("We currently have %s supply drops on screen"%(len(game_state.supply_drops)))
b2.writeLog(game_state.logs, filename = 'cbss_farming_2')
```
